```python
# --- CRÉDITOS DEL PROGRAMA ---
print("Programa realizado por Nicole Robles NL = 50")
print("Bienvenido al Cerebro Lógico del Coreógrafo IA")

# 1. Inicialización de la lista vacía (Sources [1])
historial_rutinas = []
ejecutar_sistema = "si"

# 2. Ciclo principal para el menú interactivo (Sources [2])
while ejecutar_sistema == "si":
    print("\n--- MENÚ DE IA PARA BAILARINES ---")
    print("1. Evaluar Perfil y Sugerir Rutina")
    print("2. Ver Historial de Rutinas Sugeridas")
    print("3. Borrar Base de Datos (Limpiar Lista)")
    print("4. Salir")

    opcion = input("\nSeleccione una opción (1-4): ")

    # MODO 1: Lógica de IA para evaluación (Conversación previa)
    if opcion == "1":
        print("\n[MODO IA: PROCESANDO PERFIL]")
        nombre = input("Nombre del bailarín: ")
        experiencia = int(input("Años de trayectoria: "))
        energia = float(input("Nivel de energía actual (0.0 a 10.0): "))
        ritmo = input("¿Prefiere ritmo 'rapido' o 'suave'?: ").lower()

        # Procesamiento lógico con operadores (Sources [3, 4])
        if ritmo == "rapido" and energia >= 7.5:
            recomendacion = "Breakdance (Perfil de Alto Impacto)"
        elif ritmo == "rapido" and energia < 7.5:
            recomendacion = "Tap Dance (Perfil de Energía Moderada)"
        elif ritmo == "suave" or experiencia > 10:
            recomendacion = "Ballet Lírico (Perfil de Técnica Clásica)"
        else:
            recomendacion = "Jazz Básico (Perfil en Desarrollo)"

        print(f"IA: {nombre}, la rutina recomendada es: {recomendacion}")

        # Agregar elemento a la lista (Sources [1])
        historial_rutinas.append(f"{nombre}: {recomendacion}")
        print("IA: Resultado guardado en el historial.")

    # MODO 2: Mostrar el contenido de la lista sin ciclo for
    elif opcion == "2":
        print("\n--- BASE DE DATOS DE RUTINAS ---")
        # Verificar si la lista está vacía (Sources [1, 5])
        if not historial_rutinas:
            print("IA: No hay rutinas registradas todavía.")
        else:
            # Uso de while con contador para recorrer la lista (Sources [6])
            indice = 0
            # len() devuelve la longitud de la lista (Sources [7])
            while indice < len(historial_rutinas):
                print(f"{indice + 1}. {historial_rutinas[indice]}")
                indice = indice + 1

    # MODO 3: Limpiar la lista (Sources [1, 8])
    elif opcion == "3":
        historial_rutinas = []
        print("\nIA: Base de datos reiniciada con éxito.")

    # MODO 4: Salida del sistema
    elif opcion == "4":
        ejecutar_sistema = "no"
        print("\nIA: Apagando sistema... ¡Sigue bailando!")

    # Manejo de entradas no válidas
    else:
        print("\nIA: Error. Opción no válida en el menú.")

print("\n--- Ejecución Finalizada ---")
print("Programa realizado por Nicole Robles NL = 50")
```
