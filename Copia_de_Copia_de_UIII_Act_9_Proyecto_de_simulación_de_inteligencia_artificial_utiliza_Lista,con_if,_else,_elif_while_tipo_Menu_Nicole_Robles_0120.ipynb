```python
# --- CRÉDITOS DEL PROGRAMA ---
print("Bienvenido al Cerebro Lógico de IA para Bailarines")

# 1. Inicialización de la lista vacía según lo solicitado
# Las listas en Python permiten almacenar colecciones de datos [1, 2].
historial_bailarin = []

# 2. Ciclo infinito para el menú interactivo
# El ciclo 'while True' mantiene el sistema activo hasta que se use 'break' [3, 4].
while True:
    print("\n--- MENÚ PRINCIPAL DE IA ---")
    print("1. Analizar Perfil de Bailarín")
    print("2. Ver Historial de Recomendaciones (Lista)")
    print("3. Salir")

    opcion = input("\nSeleccione una opción (1-3): ")

    # MODO 1: Lógica de IA para evaluación de talento
    if opcion == "1":
        print("\n[MODO IA: PROCESANDO DATOS]")
        nombre = input("Nombre del bailarín: ")
        experiencia = int(input("Años de trayectoria: "))
        energia = float(input("Nivel de energía (0.0 a 10.0): "))
        ritmo = input("¿Prefiere ritmo 'rapido' o 'suave'?: ").lower()

        # Motor de Decisión de la IA (Estructuras selectivas)
        # Se utilizan operadores lógicos (and, or) para simular inteligencia [5, 6].
        if ritmo == "rapido" and energia >= 7.5:
            resultado = "Rutina de Breakdance (Alto Impacto)"
        elif ritmo == "rapido" and energia < 7.5:
            resultado = "Rutina de Tap Dance (Energía Moderada)"
        elif ritmo == "suave" or experiencia > 10:
            resultado = "Ballet Lírico (Técnica Clásica)"
        else:
            resultado = "Jazz Básico (Perfil en Desarrollo)"

        print(f"IA: {nombre}, la recomendación es: {resultado}")

        # Guardar en la lista usando append() [2]
        # Almacenamos una cadena con la información procesada
        registro = "Bailarín: " + nombre + " | Sugerencia: " + resultado
        historial_bailarin.append(registro)
        print("IA: Análisis guardado en la lista correctamente.")

    # MODO 2: Mostrar el historial utilizando while (Sin ciclo for)
    elif opcion == "2":
        print("\n--- HISTORIAL DE EVALUACIONES (Contenido de la Lista) ---")

        # Verificar si la lista está vacía [2, 7]
        if not historial_bailarin:
            print("IA: La lista está vacía actualmente.")
        else:
            # Recorrido de la lista mediante un contador y while [8-10]
            indice = 0
            # len() nos da la longitud actual de la lista [10]
            while indice < len(historial_bailarin):
                print(f"{indice + 1}. {historial_bailarin[indice]}")
                indice = indice + 1

    # MODO 3: Salida del sistema
    elif opcion == "3":
        print("\nIA: Apagando cerebro lógico... ¡Sigue practicando!")
        break # Rompe el ciclo while True para finalizar [3, 4]

    # Manejo de entradas no válidas
    else:
        print("\nIA: Error. Opción no válida. Intente de nuevo.")

print("\n--- Ejecución Finalizada ---")
print("Programa realizado por Nicole Robles NL = 50")
```
