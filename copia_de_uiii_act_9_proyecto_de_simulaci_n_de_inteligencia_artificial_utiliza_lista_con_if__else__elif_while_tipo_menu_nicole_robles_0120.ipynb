```python
# Lista vacía
bailarines = []

while True:

    print("\n===== MENU BAILARIN =====")
    print("1. Agregar bailarín")
    print("2. Leer bailarín")
    print("3. Listar bailarines")
    print("4. Buscar bailarín")
    print("5. Modificar bailarín")
    print("6. Borrar bailarín")
    print("7. Salir")

    opcion = input("Selecciona una opción: ")

    # 1. Agregar bailarín
    if opcion == "1":
        nombre = input("Escribe el nombre del bailarín: ")
        bailarines.append(nombre)
        print("Bailarín agregado correctamente")

    # 2. Leer bailarín
    elif opcion == "2":

        if len(bailarines) == 0:
            print("La lista está vacía")

        else:
            try:
                posicion = int(input("Número de posición: "))
                if posicion >= 0 and posicion < len(bailarines):
                    print("Bailarín encontrado:", bailarines[posicion])
                else:
                    print("Posición incorrecta")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la posición.")

    # 3. Listar bailarines
    elif opcion == "3":

        if len(bailarines) == 0:
            print("No hay bailarines registrados")

        else:
            i = 0

            while i < len(bailarines):
                print(i, "-", bailarines[i])
                i = i + 1

    # 4. Buscar bailarín
    elif opcion == "4":

        buscar = input("Nombre del bailarín a buscar: ")

        if buscar in bailarines:
            print("Sí existe el bailarín")

        else:
            print("No se encontró el bailarín")

    # 5. Modificar bailarín
    elif opcion == "5":

        if len(bailarines) == 0:
            print("La lista está vacía")

        else:
            try:
                posicion = int(input("Posición a modificar: "))
                if posicion >= 0 and posicion < len(bailarines):

                    nuevo = input("Nuevo nombre: ")
                    bailarines[posicion] = nuevo

                    print("Bailarín modificado")

                else:
                    print("Posición incorrecta")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la posición.")

    # 6. Borrar bailarín
    elif opcion == "6":

        if len(bailarines) == 0:
            print("La lista está vacía")

        else:
            try:
                posicion = int(input("Posición a borrar: "))

                if posicion >= 0 and posicion < len(bailarines):

                    eliminado = bailarines.pop(posicion)

                    print("Bailarín eliminado:", eliminado)

                else:
                    print("Posición incorrecta")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la posición.")

    # 7. Salir
    elif opcion == "7":
        print("/////Lista/////")
        print("\nDatos guardados correctamente")
        print("Programa terminado")
        break

    # Opción inválida
    else:
        print("Opción no válida")
        print("Programa Realizado por Nicole Robles Mendoza")
```
